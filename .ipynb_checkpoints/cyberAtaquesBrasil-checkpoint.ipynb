```python
# Numero de ataque por ano definido
def plot_attacks_in_year(data, year):
    valid_attacks = ['Worm', 'DOS', 'Invasao', 'Web', 'Scan', 'Fraude', 'Outros']
    
    # Verificar se o ano fornecido está presente nos dados
    if year not in data['Ano'].unique():
        raise ValueError(f"Ano inválido. Escolha entre os anos disponíveis: {', '.join(map(str, data['Ano'].unique()))}")
    
    # Filtra os dados para o ano escolhido
    year_data = data[data['Ano'] == year]
    
    # Soma o número de incidentes por tipo de ataque
    attack_sums = year_data[valid_attacks].sum().reset_index()
    attack_sums.columns = ['Tipo de Ataque', 'Número de Incidentes']

    # Plotar o gráfico
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Tipo de Ataque', y='Número de Incidentes', data=attack_sums, palette='viridis')
    plt.title(f'Número de Incidentes por Tipo de Ataque em {year}')
    plt.xlabel('Tipo de Ataque')
    plt.ylabel('Número de Incidentes')
    plt.xticks(rotation=45)
    plt.show()


# variavel criada para mudança em código do ano, na anaálise de ataque por ano
year = 2018 

plot_attacks_in_year(data, year)
```

```python
# Lista de tipos de incidentes disponíveis
incident_types = ['Worm', 'DOS', 'Invasao', 'Web', 'Scan', 'Fraude', 'Outros']

def plot_attack_by_year(attack_type):
    if attack_type not in incident_types:
        print(f"Tipo de ataque inválido. Por favor, escolha entre: {', '.join(incident_types)}")
        return

    plt.figure(figsize=(10, 6))
    sns.lineplot(x='Ano', y=attack_type, data=data, marker='o')
    plt.title(f'Incidência de {attack_type} por Ano')
    plt.xlabel('Ano')
    plt.ylabel(f'Número de Incidentes de {attack_type}')
    plt.grid(True)
    plt.show()

# variavel criada para mudança em código do ataque, na análise de crescimento por cada tipo de ataque
#usar o array inical como referencia de parâmetro da função
plot_attack_by_year('Outros')
```

```python
# Evolução mensal de cada incidente pelos anos
# usar o grafico antior de exemplo, podendo mudar o tipo de ataque no array abaixo
select_incident = "Scan"
mensal_incidentes = data.groupby(['Ano', 'Mes'])[select_incident].sum().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(x='Mes', y=select_incident, hue='Ano', data=mensal_incidentes, marker='o', palette='viridis')
plt.title('Incidentes '+select_incident+' ao Longo dos Meses e Anos')
plt.xlabel('Mês')
plt.ylabel('Número de Incidentes '+select_incident+'')
plt.xticks(rotation=45)
plt.legend(title='Ano')
plt.show()
```

```python
# Calcular a matriz de correlação entre os ataque, utilizado o método de Pearson por default
# outro exemplo method='spearman'
corr_matrix = data[incident_types].corr()

# Visualizar a matriz de correlação com um heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlação entre Tipos de Incidentes')
plt.show()
```

```python
# Calcular a proporção de cada tipo de incidente
proportions = data[incident_types].sum() / data[incident_types].sum().sum()

# Plotar as proporções
proportions.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proporção de Cada Tipo de Incidente')
plt.ylabel('')
plt.show()
```

```python
# Plotar o crescimento de incidentes por categoria ao longo dos anos
plt.figure(figsize=(14, 7))
for incident in incident_types:
    sns.lineplot(x='Ano', y=incident, data=data, marker='o', label=incident)

plt.title('Crescimento de Incidentes por Categoria ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Incidentes')
plt.legend(title='Tipo de Incidente')
plt.show()
```

```python
# Criar gráficos de distribuição para cada tipo de incidente ao longo dos meses
meses_ordem = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 
               'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

# escolha aqui o tipo de ataque que deseja ver
incident = "Worm"

plt.figure(figsize=(16, 12))
plt.subplot(4, 2, i)
sns.boxplot(x='Mes', y=incident, data=data, order=meses_ordem)
plt.title(f'Distribuição Mensal de Incidentes: {incident}')
plt.xlabel('Mês')
plt.ylabel('Número de Incidentes')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Criar um gráfico de linha para visualizar as tendências mensais
data['Mes'] = pd.Categorical(data['Mes'], categories=meses_ordem, ordered=True)

plt.figure(figsize=(14, 7))
sns.lineplot(x='Mes', y='Total', hue='Ano', data=data, marker='o', palette='tab10')
plt.title('Tendências Mensais de Incidentes ao Longo dos Anos')
plt.xlabel('Mês')
plt.ylabel('Total de Incidentes')
plt.xticks(rotation=45)
plt.legend(title='Ano')
plt.show()
```

```python
# número total de ataques por ano

total_por_ano = data.groupby('Ano')['Total'].sum()

plt.figure(figsize=(12, 6))
plt.plot(total_por_ano.index, total_por_ano.values, marker='o', linestyle='-', color='b')
plt.title('Número Total de Ataques por Ano')
plt.xlabel('Ano')
plt.ylabel('Total de Ataques')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de número total de ataques por mês
total_por_mes = data.groupby('Mes')['Total'].sum()
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
         'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
plt.figure(figsize=(12, 6))
plt.bar(meses, total_por_mes.values, color='skyblue')
plt.title('Número Total de Ataques por Mês (Todos os Anos)')
plt.xlabel('Mês')
plt.ylabel('Total de Ataques')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
